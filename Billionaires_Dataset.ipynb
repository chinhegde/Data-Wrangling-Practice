```python
# What are the unique values in the 'nationality' column?

df['nationality'] = df['nationality'].str.replace(r'[^\x00-\x7F]+', '', regex = True)
df['nationality'].unique()
```

```python
# How many unique values are in the 'source_of_wealth' column?

df['source_of_wealth'] = df['source_of_wealth'].str.replace(r'[^\x00-\x7F]+', '', regex = True)
df['source_of_wealth'].nunique()
```

```python
# Get the summary statistics of the DataFrame.

df['yearly_ranking'] = df['yearly_ranking'].astype('int64')
df.describe(include='all')
```

```python
# How many entries are there for each 'year'?

df.groupby(by='year').count()['name']
```

```python
# How can you sort the DataFrame by net_worth_usd in descending order?

df[['name', 'net_worth_usd']].sort_values('net_worth_usd', ascending = False)
```

```python
# How can you add a new column age_group with values 'Young', 'Middle-aged', 'Old' based on the age column?

df['age'].unique()
df.drop(df[~df['age'].str.isdigit()].index, inplace = True)

df['age'] = df['age'].astype('int64')
```

```python
df['age'].min(), df['age'].max()
```

```python
df['age_group'] = pd.cut(df['age'], bins=[0, 40, 60, float('inf')], labels=['Young', 'Middle-aged', 'Old'], right=False)
```

```python
df['age_group'].value_counts()
```

## Advanced questions

```python
# How can you create a function that normalizes the net_worth_usd column and apply it to the DataFrame?

def normalize(col):
  return (col - col.min())/(col.max() - col.min())

df['net_worth_usd'] = df['net_worth_usd'].str.replace('billion', '')
df['net_worth_usd'] = df['net_worth_usd'].str.replace('$', '')
df['net_worth_usd'] = df['net_worth_usd'].str.replace(' ', '')

df['net_worth_usd'] = (df['net_worth_usd'].astype('float64'))*1000000000

df['normalized_net_worth'] = normalize(df['net_worth_usd'])
```
